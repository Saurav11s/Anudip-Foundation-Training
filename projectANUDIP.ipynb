```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
url = "https://raw.githubusercontent.com/AnudipAE/DANLC/master/cleaned.csv"
df = pd.read_csv(url)

df.head()
```

```python

#  Data Understanding
df.shape
df.info()
df.describe()
```

```python
#  Data Cleaning
df.isnull().sum()
```

```python

#  Exploratory Data Analysis (EDA)

df.groupby('year')['amount'].count().plot(kind='bar', title='Year-wise Sales')
plt.show()
```

```python


df_15_18 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

df_15_18.groupby('month')['rating'].count().plot(kind='bar')
plt.show()
```

```python


df_15_18.groupby('brand')['amount'].sum() \
.sort_values(ascending=False).head(10) \
.plot(kind='bar', title='Top Brands')
plt.show()
```

```python

df_15_18.groupby('category')['amount'].sum() \
.sort_values(ascending=False).head(10) \
.plot(kind='bar', title='Top Categories')
plt.show()
```

```python
sns.countplot(x='rating', data=df)
plt.show()
```

```python


df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
Conclusion
2015 was the best year in terms of overall sales performance.
January recorded the highest sales among all months.
Headphones were the top-selling category, followed by Computers & Accessories.
Bose and Logitech were the leading brands with maximum sales.
Archos and EINCAR had the lowest sales performance.
Most products received a rating of 5, indicating high customer satisfaction.
Savage and Plemo were the highest-rated brands.
Female customers contributed the majority of purchases.
Sales showed steady growth until 2015, followed by a decline in later years.
```
